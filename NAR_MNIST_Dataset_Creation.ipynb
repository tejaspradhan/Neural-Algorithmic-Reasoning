```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
dataset = []
for i in range(len(x_train)):
  dataset.append([x_train[i],y_train[i]])
```

```python
dataset = np.array(dataset)
```

```python
def create_image_data(num_instances, dataset):
  i=0
  x_train_data= []
  y_train_data = []
  while i<len(dataset):
    image_data_x = dataset[i:i+num_instances] # stacking num_instances amount of images and labels together

    x_vals = []
    y_vals = []
    before_sorting = []
    after_sorting = []

    # separating unordered images and storing into x training variables
    for element in image_data_x:
      x_vals.append(element[0])
      before_sorting.append(element[1])
    x_train_data.append(x_vals)

    after_sorting = before_sorting.copy()
    after_sorting.sort()

    for element in after_sorting:
      y_vals.append(before_sorting.index(element))

    # sorting images according to their labels
    # image_data_x = list(image_data_x)
    # image_data_x.sort(key=lambda x: x[1])
    image_data_x = np.array(image_data_x)

    # storing ordered images as y training labels
    # for element in image_data_x:
      # y_vals.append(element[0])
      # print(element[1])
    y_train_data.append(y_vals)

    
    print(np.array(x_train_data).shape)
    i+=num_instances
  x_train_data = np.array(x_train_data)
  y_train_data = np.array(y_train_data)
  return x_train_data,y_train_data
    
```
